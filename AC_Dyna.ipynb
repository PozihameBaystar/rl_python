```python
"""
深層学習版Dynaアルゴリズムの完全実装
連続状態空間・連続行動空間に対応したモデルベース強化学習

環境: Gymnasium Pendulum-v1
アプローチ: Actor-Critic + Model-based Planning
"""

import gymnasium as gym
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from collections import deque
import random
import matplotlib.pyplot as plt

# ============================================
# 初期設定
# ============================================
SEED = 42
torch.manual_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"使用デバイス: {device}")

# GPUが利用可能な場合、GPU情報を表示
if torch.cuda.is_available():
    print(f"GPU名: {torch.cuda.get_device_name(0)}")
    print(f"総メモリ: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.2f} GB")
    print("-" * 50)


# ============================================
# GPU監視ユーティリティ
# ============================================

class GPUMonitor:
    """
    GPU使用状況を監視するクラス
    PyTorchの組み込み機能を使ってメモリ使用量などを追跡します
    """
    def __init__(self):
        self.is_cuda_available = torch.cuda.is_available()
        if self.is_cuda_available:
            self.device_name = torch.cuda.get_device_name(0)
            self.total_memory = torch.cuda.get_device_properties(0).total_memory
    
    def get_memory_info(self):
        """
        現在のGPUメモリ使用状況を取得
        
        返り値:
            辞書形式のメモリ情報（MB単位）
        """
        if not self.is_cuda_available:
            return {
                'allocated': 0,
                'reserved': 0,
                'free': 0,
                'total': 0,
                'utilization': 0
            }
        
        # 割り当て済みメモリ（実際にテンソルが使用中）
        allocated = torch.cuda.memory_allocated(0) / 1024**2  # MB
        
        # 予約済みメモリ（CUDAメモリキャッシュ含む）
        reserved = torch.cuda.memory_reserved(0) / 1024**2  # MB
        
        # 総メモリ
        total = self.total_memory / 1024**2  # MB
        
        # 空きメモリ
        free = total - reserved
        
        # 使用率（%）
        utilization = (reserved / total) * 100 if total > 0 else 0
        
        return {
            'allocated': allocated,
            'reserved': reserved,
            'free': free,
            'total': total,
            'utilization': utilization
        }
    
    def print_memory_info(self, prefix=""):
        """
        GPUメモリ情報を整形して表示
        
        引数:
            prefix: 表示の前に付ける文字列
        """
        if not self.is_cuda_available:
            print(f"{prefix}GPU: 利用不可（CPUモード）")
            return
        
        info = self.get_memory_info()
        
        print(f"{prefix}GPU使用状況:")
        print(f"{prefix}  使用中: {info['allocated']:.1f} MB")
        print(f"{prefix}  予約済: {info['reserved']:.1f} MB")
        print(f"{prefix}  空き: {info['free']:.1f} MB")
        print(f"{prefix}  合計: {info['total']:.1f} MB")
        print(f"{prefix}  利用率: {info['utilization']:.1f}%")
    
    def reset_peak_memory_stats(self):
        """ピークメモリ統計をリセット"""
        if self.is_cuda_available:
            torch.cuda.reset_peak_memory_stats(0)
    
    def get_peak_memory_info(self):
        """
        訓練中のピークメモリ使用量を取得
        
        返り値:
            ピークメモリ情報（MB単位）
        """
        if not self.is_cuda_available:
            return {'peak_allocated': 0, 'peak_reserved': 0}
        
        peak_allocated = torch.cuda.max_memory_allocated(0) / 1024**2
        peak_reserved = torch.cuda.max_memory_reserved(0) / 1024**2
        
        return {
            'peak_allocated': peak_allocated,
            'peak_reserved': peak_reserved
        }


# グローバルなGPUモニターインスタンスを作成
gpu_monitor = GPUMonitor()


# ============================================
# ニューラルネットワークの定義
# ============================================

class ModelNetwork(nn.Module):
    """
    環境のダイナミクスを学習するネットワーク
    入力: 状態sと行動a
    出力: 次状態の差分Δs = s' - s と報酬r
    """
    def __init__(self, state_dim, action_dim, hidden_dim=256):
        super(ModelNetwork, self).__init__()
        
        input_dim = state_dim + action_dim
        
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        self.fc3 = nn.Linear(hidden_dim, hidden_dim)
        
        self.delta_state_head = nn.Linear(hidden_dim, state_dim)
        self.reward_head = nn.Linear(hidden_dim, 1)
        
    def forward(self, state, action):
        x = torch.cat([state, action], dim=-1)
        
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        
        delta_state = self.delta_state_head(x)
        reward = self.reward_head(x)
        
        return delta_state, reward
    
    def predict_next_state(self, state, action):
        """次状態を予測する便利メソッド"""
        delta_state, reward = self.forward(state, action)
        next_state = state + delta_state
        return next_state, reward


class CriticNetwork(nn.Module):
    """
    行動価値関数Q(s,a)を近似するネットワーク
    入力: 状態sと行動a
    出力: Q値（スカラー）
    """
    def __init__(self, state_dim, action_dim, hidden_dim=256):
        super(CriticNetwork, self).__init__()
        
        input_dim = state_dim + action_dim
        
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        self.fc3 = nn.Linear(hidden_dim, hidden_dim)
        self.q_head = nn.Linear(hidden_dim, 1)
        
    def forward(self, state, action):
        x = torch.cat([state, action], dim=-1)
        
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        
        q_value = self.q_head(x)
        return q_value


class ActorNetwork(nn.Module):
    """
    方策π(s)を近似するネットワーク
    入力: 状態s
    出力: 行動a（連続値）
    """
    def __init__(self, state_dim, action_dim, action_max, hidden_dim=256):
        super(ActorNetwork, self).__init__()
        
        self.action_max = action_max
        
        self.fc1 = nn.Linear(state_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        self.fc3 = nn.Linear(hidden_dim, hidden_dim)
        self.action_head = nn.Linear(hidden_dim, action_dim)
        
    def forward(self, state):
        x = F.relu(self.fc1(state))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        
        action = torch.tanh(self.action_head(x)) * self.action_max
        return action


# ============================================
# リプレイバッファ
# ============================================

class ReplayBuffer:
    """経験を保存し、ランダムサンプリングするためのバッファ"""
    def __init__(self, capacity=100000):
        self.buffer = deque(maxlen=capacity)
    
    def push(self, state, action, reward, next_state, done):
        """経験をバッファに追加"""
        self.buffer.append((state, action, reward, next_state, done))
    
    def sample(self, batch_size):
        """バッファからランダムにbatch_size個の経験をサンプリング"""
        batch = random.sample(self.buffer, batch_size)
        states, actions, rewards, next_states, dones = zip(*batch)
        
        states = torch.FloatTensor(np.array(states)).to(device)
        actions = torch.FloatTensor(np.array(actions)).to(device)
        rewards = torch.FloatTensor(np.array(rewards)).unsqueeze(1).to(device)
        next_states = torch.FloatTensor(np.array(next_states)).to(device)
        dones = torch.FloatTensor(np.array(dones)).unsqueeze(1).to(device)
        
        return states, actions, rewards, next_states, dones
    
    def __len__(self):
        return len(self.buffer)


# ============================================
# Dynaエージェント
# ============================================

class DynaAgent:
    """深層学習版のDynaエージェント"""
    def __init__(
        self,
        state_dim,
        action_dim,
        action_max,
        lr_actor=3e-4,
        lr_critic=3e-4,
        lr_model=1e-3,
        gamma=0.99,
        tau=0.005,
        buffer_size=100000,
        batch_size=256,
        n_planning=5
    ):
        self.state_dim = state_dim
        self.action_dim = action_dim
        self.action_max = action_max
        self.gamma = gamma
        self.tau = tau
        self.batch_size = batch_size
        self.n_planning = n_planning
        
        # ネットワークの初期化
        self.actor = ActorNetwork(state_dim, action_dim, action_max).to(device)
        self.actor_target = ActorNetwork(state_dim, action_dim, action_max).to(device)
        self.actor_target.load_state_dict(self.actor.state_dict())
        
        self.critic = CriticNetwork(state_dim, action_dim).to(device)
        self.critic_target = CriticNetwork(state_dim, action_dim).to(device)
        self.critic_target.load_state_dict(self.critic.state_dict())
        
        self.model = ModelNetwork(state_dim, action_dim).to(device)
        
        # オプティマイザの初期化
        self.actor_optimizer = optim.Adam(self.actor.parameters(), lr=lr_actor)
        self.critic_optimizer = optim.Adam(self.critic.parameters(), lr=lr_critic)
        self.model_optimizer = optim.Adam(self.model.parameters(), lr=lr_model)
        
        # リプレイバッファ
        self.replay_buffer = ReplayBuffer(buffer_size)
    
    def select_action(self, state, add_noise=True, noise_scale=0.1):
        """状態を受け取って行動を選択"""
        state = torch.FloatTensor(state).unsqueeze(0).to(device)
        
        with torch.no_grad():
            action = self.actor(state).cpu().numpy()[0]
        
        if add_noise:
            noise = np.random.normal(0, noise_scale, size=self.action_dim)
            action = action + noise
            action = np.clip(action, -self.action_max, self.action_max)
        
        return action
    
    def update_critic(self, states, actions, rewards, next_states, dones):
        """Criticネットワークを更新"""
        with torch.no_grad():
            next_actions = self.actor_target(next_states)
            target_q = self.critic_target(next_states, next_actions)
            target_q = rewards + (1 - dones) * self.gamma * target_q
        
        current_q = self.critic(states, actions)
        critic_loss = F.mse_loss(current_q, target_q)
        
        self.critic_optimizer.zero_grad()
        critic_loss.backward()
        self.critic_optimizer.step()
        
        return critic_loss.item()
    
    def update_actor(self, states):
        """Actorネットワークを更新"""
        actions = self.actor(states)
        q_values = self.critic(states, actions)
        actor_loss = -q_values.mean()
        
        self.actor_optimizer.zero_grad()
        actor_loss.backward()
        self.actor_optimizer.step()
        
        return actor_loss.item()
    
    def update_model(self, states, actions, rewards, next_states):
        """Modelネットワークを更新"""
        predicted_delta_state, predicted_reward = self.model(states, actions)
        
        actual_delta_state = next_states - states
        
        state_loss = F.mse_loss(predicted_delta_state, actual_delta_state)
        reward_loss = F.mse_loss(predicted_reward, rewards)
        
        model_loss = state_loss + reward_loss
        
        self.model_optimizer.zero_grad()
        model_loss.backward()
        self.model_optimizer.step()
        
        return model_loss.item()
    
    def planning(self):
        """
        学習済みのModelを使ってシミュレーションを行い、
        架空の経験からActor-Criticを追加学習（Dynaの核心部分）
        """
        if len(self.replay_buffer) < self.batch_size:
            return
        
        for _ in range(self.n_planning):
            states, actions, _, _, _ = self.replay_buffer.sample(self.batch_size)
            
            # Modelを使って次状態と報酬をシミュレート
            with torch.no_grad():
                simulated_next_states, simulated_rewards = self.model.predict_next_state(states, actions)
            
            simulated_dones = torch.zeros_like(simulated_rewards)
            
            # Criticの更新
            with torch.no_grad():
                next_actions = self.actor_target(simulated_next_states)
                target_q = self.critic_target(simulated_next_states, next_actions)
                target_q = simulated_rewards + self.gamma * target_q
            
            current_q = self.critic(states, actions)
            critic_loss = F.mse_loss(current_q, target_q)
            
            self.critic_optimizer.zero_grad()
            critic_loss.backward()
            self.critic_optimizer.step()
            
            # Actorの更新
            simulated_actions = self.actor(states)
            simulated_next_states_for_actor, _ = self.model.predict_next_state(states, simulated_actions)
            actor_loss = -self.critic(simulated_next_states_for_actor, simulated_actions).mean()
            
            self.actor_optimizer.zero_grad()
            actor_loss.backward()
            self.actor_optimizer.step()
    
    def update_target_networks(self):
        """ターゲットネットワークをソフト更新"""
        for target_param, param in zip(self.actor_target.parameters(), self.actor.parameters()):
            target_param.data.copy_(self.tau * param.data + (1 - self.tau) * target_param.data)
        
        for target_param, param in zip(self.critic_target.parameters(), self.critic.parameters()):
            target_param.data.copy_(self.tau * param.data + (1 - self.tau) * target_param.data)
    
    def train_step(self):
        """
        一回の学習ステップを実行
        直接強化学習→モデル学習→プランニングの流れを実装
        """
        if len(self.replay_buffer) < self.batch_size:
            return None
        
        # リプレイバッファから経験をサンプリング
        states, actions, rewards, next_states, dones = self.replay_buffer.sample(self.batch_size)
        
        # 直接強化学習（実際の経験からActor-Criticを更新）
        critic_loss = self.update_critic(states, actions, rewards, next_states, dones)
        actor_loss = self.update_actor(states)
        
        # モデル学習（実際の遷移からModelを学習）
        model_loss = self.update_model(states, actions, rewards, next_states)
        
        # プランニング（学習したModelを使って架空の経験から追加学習）
        self.planning()
        
        # ターゲットネットワークの更新
        self.update_target_networks()
        
        return {
            'critic_loss': critic_loss,
            'actor_loss': actor_loss,
            'model_loss': model_loss
        }


# ============================================
# 訓練関数
# ============================================

def train_dyna_agent(
    env_name='Pendulum-v1',
    n_episodes=200,
    max_steps=200,
    n_planning=10,
    warmup_steps=1000,
    noise_scale=0.1,
    noise_decay=0.9995,
    show_gpu_every=10  # 何エピソードごとにGPU情報を表示するか
):
    """Dynaエージェントを訓練するメイン関数"""
    env = gym.make(env_name)
    
    state_dim = env.observation_space.shape[0]
    action_dim = env.action_space.shape[0]
    action_max = env.action_space.high[0]
    
    print(f"環境: {env_name}")
    print(f"状態次元: {state_dim}, 行動次元: {action_dim}")
    print(f"プランニングステップ数: {n_planning}")
    print("-" * 50)
    
    # 初期GPU状態を表示
    gpu_monitor.print_memory_info("初期状態 - ")
    print("-" * 50)
    
    # ピークメモリ統計をリセット
    gpu_monitor.reset_peak_memory_stats()
    
    agent = DynaAgent(
        state_dim=state_dim,
        action_dim=action_dim,
        action_max=action_max,
        n_planning=n_planning
    )
    
    episode_rewards = []
    total_steps = 0
    current_noise_scale = noise_scale
    
    for episode in range(n_episodes):
        state, _ = env.reset(seed=SEED + episode)
        episode_reward = 0
        episode_losses = {'critic_loss': 0, 'actor_loss': 0, 'model_loss': 0}
        episode_steps = 0
        
        for step in range(max_steps):
            # 行動選択
            if total_steps < warmup_steps:
                action = env.action_space.sample()
            else:
                action = agent.select_action(state, add_noise=True, noise_scale=current_noise_scale)
            
            # 環境で行動を実行
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            
            episode_reward += reward
            episode_steps += 1
            total_steps += 1
            
            # 経験をリプレイバッファに保存
            agent.replay_buffer.push(state, action, reward, next_state, done)
            
            # 学習の実行
            if total_steps >= warmup_steps:
                losses = agent.train_step()
                
                if losses is not None:
                    for key, value in losses.items():
                        episode_losses[key] += value
            
            state = next_state
            
            if done:
                break
        
        episode_rewards.append(episode_reward)
        current_noise_scale *= noise_decay
        
        if episode_steps > 0:
            for key in episode_losses:
                episode_losses[key] /= episode_steps
        
        if (episode + 1) % 10 == 0:
            recent_rewards = episode_rewards[-10:]
            avg_reward = np.mean(recent_rewards)
            print(f"エピソード {episode + 1}/{n_episodes}")
            print(f"  平均報酬（直近10エピソード）: {avg_reward:.2f}")
            print(f"  このエピソードの報酬: {episode_reward:.2f}")
            print(f"  Critic損失: {episode_losses['critic_loss']:.4f}")
            print(f"  Actor損失: {episode_losses['actor_loss']:.4f}")
            print(f"  Model損失: {episode_losses['model_loss']:.4f}")
            print(f"  探索ノイズ: {current_noise_scale:.4f}")
            print(f"  総ステップ数: {total_steps}")
            
            # GPU使用状況を表示
            if (episode + 1) % show_gpu_every == 0:
                gpu_monitor.print_memory_info("  ")
            
            print("-" * 50)
    
    # 訓練終了後のGPU状態とピークメモリを表示
    print("\n訓練完了後のGPU状態:")
    gpu_monitor.print_memory_info()
    
    if gpu_monitor.is_cuda_available:
        peak_info = gpu_monitor.get_peak_memory_info()
        print(f"\n訓練中のピークメモリ使用量:")
        print(f"  ピーク使用中: {peak_info['peak_allocated']:.1f} MB")
        print(f"  ピーク予約済: {peak_info['peak_reserved']:.1f} MB")
    
    print("-" * 50)
    
    env.close()
    return agent, episode_rewards


# ============================================
# 可視化関数
# ============================================

def plot_training_results(episode_rewards, window_size=10):
    """訓練結果を可視化する関数"""
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(episode_rewards, alpha=0.3, label='エピソード報酬')
    
    if len(episode_rewards) >= window_size:
        moving_avg = np.convolve(episode_rewards, np.ones(window_size)/window_size, mode='valid')
        plt.plot(range(window_size-1, len(episode_rewards)), moving_avg, 
                label=f'{window_size}エピソード移動平均', linewidth=2)
    
    plt.xlabel('エピソード')
    plt.ylabel('累積報酬')
    plt.title('訓練の進捗')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(1, 2, 2)
    plt.hist(episode_rewards, bins=30, edgecolor='black', alpha=0.7)
    plt.xlabel('累積報酬')
    plt.ylabel('頻度')
    plt.title('報酬の分布')
    plt.axvline(np.mean(episode_rewards), color='red', linestyle='--', 
                linewidth=2, label=f'平均: {np.mean(episode_rewards):.2f}')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()


# ============================================
# テスト関数
# ============================================

def test_agent(agent, env_name='Pendulum-v1', n_episodes=5, render=False):
    """学習済みエージェントをテストする関数"""
    if render:
        env = gym.make(env_name, render_mode='human')
    else:
        env = gym.make(env_name)
    
    test_rewards = []
    
    print("\n" + "="*50)
    print("学習済みエージェントのテスト")
    print("="*50)
    
    for episode in range(n_episodes):
        state, _ = env.reset()
        episode_reward = 0
        done = False
        steps = 0
        
        while not done and steps < 200:
            action = agent.select_action(state, add_noise=False)
            state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            episode_reward += reward
            steps += 1
        
        test_rewards.append(episode_reward)
        print(f"テストエピソード {episode + 1}: 報酬 = {episode_reward:.2f}")
    
    avg_reward = np.mean(test_rewards)
    std_reward = np.std(test_rewards)
    
    print("-"*50)
    print(f"平均報酬: {avg_reward:.2f} ± {std_reward:.2f}")
    print("="*50)
    
    env.close()
    return avg_reward


# ============================================
# メイン実行部分
# ============================================

if __name__ == "__main__":
    print("深層学習版Dynaアルゴリズムの訓練を開始します")
    print("="*50)
    
    # Dynaエージェントの訓練
    agent_dyna, rewards_dyna = train_dyna_agent(
        env_name='Pendulum-v1',
        n_episodes=200,
        n_planning=10,  # 1回の実行動あたり10回のプランニング
        warmup_steps=1000
    )
    
    # 結果の可視化
    plot_training_results(rewards_dyna)
    
    # 学習済みエージェントのテスト
    test_agent(agent_dyna, n_episodes=5)
    
    print("\n訓練が完了しました！")
    
    # オプション: プランニングなし（通常のActor-Critic）との比較
    print("\n比較のため、プランニングなしのAgent（通常のActor-Critic）も訓練します")
    print("="*50)
    
    agent_no_planning, rewards_no_planning = train_dyna_agent(
        env_name='Pendulum-v1',
        n_episodes=200,
        n_planning=0,  # プランニングなし
        warmup_steps=1000
    )
    
    # 比較プロット
    plt.figure(figsize=(10, 6))
    window_size = 10
    
    # Dynaの結果
    moving_avg_dyna = np.convolve(rewards_dyna, np.ones(window_size)/window_size, mode='valid')
    plt.plot(range(window_size-1, len(rewards_dyna)), moving_avg_dyna, 
            label='Dyna (n_planning=10)', linewidth=2)
    
    # プランニングなしの結果
    moving_avg_no_planning = np.convolve(rewards_no_planning, np.ones(window_size)/window_size, mode='valid')
    plt.plot(range(window_size-1, len(rewards_no_planning)), moving_avg_no_planning, 
            label='Actor-Critic (n_planning=0)', linewidth=2)
    
    plt.xlabel('エピソード')
    plt.ylabel('累積報酬（移動平均）')
    plt.title('DynaとActor-Criticの学習効率の比較')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    print("\n全ての訓練と比較が完了しました！")
```

```python
# ============================================
# 学習済みエージェントの可視化コード
# 訓練完了後に実行してください
# ============================================

def visualize_trained_agent(agent, env_name='Pendulum-v1', n_episodes=3, max_steps=200):
    """
    学習済みエージェントの動作を可視化する関数
    Gymnasiumのレンダリング機能を使って、実際の制御を見ることができます
    
    引数:
        agent: 学習済みのDynaエージェント
        env_name: 環境の名前
        n_episodes: 可視化するエピソード数
        max_steps: 各エピソードの最大ステップ数
    """
    print("\n" + "="*60)
    print("学習済みエージェントの動作を可視化します")
    print("ウィンドウが開き、振り子の制御を見ることができます")
    print("="*60)
    
    # render_mode='human'で環境を作成
    # これにより、リアルタイムで環境が描画されます
    env = gym.make(env_name, render_mode='human')
    
    for episode in range(n_episodes):
        print(f"\n--- エピソード {episode + 1}/{n_episodes} ---")
        
        # 環境をリセット
        state, _ = env.reset()
        episode_reward = 0
        
        print("推論開始...")
        
        for step in range(max_steps):
            # エージェントに行動を選択させる
            # add_noise=Falseで、学習した決定的な方策を使用
            action = agent.select_action(state, add_noise=False)
            
            # 環境で行動を実行
            # render_mode='human'なので、ここで自動的に描画されます
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            
            # 報酬を累積
            episode_reward += reward
            
            # 状態を更新
            state = next_state
            
            # エピソード終了判定
            if done:
                print(f"エピソード終了: {step + 1}ステップで完了")
                break
        
        print(f"累積報酬: {episode_reward:.2f}")
        
        # エピソード間で少し待機（次のエピソードとの区切りを分かりやすくするため）
        if episode < n_episodes - 1:
            print("次のエピソードまで2秒待機...")
            import time
            time.sleep(2)
    
    # 環境を閉じる
    env.close()
    print("\n" + "="*60)
    print("可視化が完了しました")
    print("="*60)


# ============================================
# 実行部分
# ============================================

# 訓練が完了し、agent_dynaが存在することを前提とします
# 3エピソード分の動作を可視化
visualize_trained_agent(agent_dyna, n_episodes=10)
```
